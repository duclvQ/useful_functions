```python
#Rz
theta=0
import math 


rotation_matrix_x=np.array([
    [1,0,0],
    [0,math.cos(theta),-math.sin(theta)],
    [0,math.sin(theta),math.cos(theta)]])
rotation_matrix_y=np.array([
    [math.cos(theta),0,math.sin(theta)],
    [0,1,0],
    [-math.sin(theta),0,math.cos(theta)]])

rotation_matrix_z=np.array([
    [math.cos(theta),-math.sin(theta),0],
    [math.sin(theta),math.cos(theta),0],
    [0,0,1]])

def transform(skel_data,transformation_matrix):
    #new_skel=[]
    new_skel_f=[]
    for v in skel_data:
        new_skel=[]
        for i in v:
            i=i.transpose()
            new_data=transformation_matrix.dot(i)
            new_data.transpose()
            new_skel.append(new_data)
        new_skel_f.append(new_skel)
    new_skel_f=np.array(new_skel_f)
    return new_skel_f
```

```python
skel_seq=np.load('skel_seq.npy')
gt_seq=np.load('gt_seq.npy')
```

```python
import math
acc_vector=[-0.106227, -0.885226, 0.466422]

c1=math.sqrt(pow(acc_vector[1],2)+pow(acc_vector[2],2))
s1=acc_vector[0]
c2=acc_vector[1]/c1 if c1 else 1.0
s2=acc_vector[2]/c1 if c1 else 0.0

rotation_matrix=np.array([
    [c1,-s1*c2,-s1*s2,0],
    [acc_vector[0],acc_vector[1],acc_vector[2],0],
    [0,-s2,c2,0],
    [0,0,0,1]])

print (rotation_matrix)

def transform(skel_data,transformation_matrix):
    new_skel=[]
    for v in skel_data:
        trans= np.transpose(v[:,:3])
        trans= np.append(trans,np.ones((1,trans.shape[1])),axis=0)

        new_data=np.matmul(transformation_matrix,trans)
        new_data=np.transpose(new_data[:3,:])
        new_skel.append(new_data)
    new_skel=np.array(new_skel)
    return new_skel

new_skel_seq=[]
for i in range(skel_seq.shape[0]):
    seq = transform(skel_seq[i],rotation_matrix)
    if seq.shape[0]==0:
        continue
    min_val = np.min(np.min(seq,axis=0),axis=0)
    max_val = np.max(np.max(seq,axis=0),axis=0)
    dist = max_val - min_val
    middle =  min_val + dist/2.0
    scale = np.max(dist)
    seq = (seq-middle)/scale
#     print seq.shape
    x,y,z=seq[:,:,0].copy(),seq[:,:,2].copy(),seq[:,:,1].copy()
    seq[:,:,0],seq[:,:,1],seq[:,:,2]=x,y,-z
    
    new_skel_seq.append(seq)
```

```python
joints_order = [v-1 for v in [1,2,2,3,3,4,3,5,5,6,6,7,7,8,3,9,9,10,10,11,11,12,1,13,13,14,14,15,15,16,1,17,17,18,18,19,19,20]]

def plot_skel(seq,num_step=5,figsize=(20,5)):   
    
    fig = plt.figure(figsize=figsize)
    step_size = int(len(seq)/float(num_step))
    for i in range(num_step):
        ax = fig.add_subplot(1, num_step, i+1, projection='3d')
#         ax = Axes3D(plot)
        skel = seq[step_size*i]
        for id1,id2 in zip(joints_order[::2],joints_order[1::2]):
            xs, ys, zs = [], [], []
            xs=[skel[id1,0],skel[id2,0]]
            ys=[skel[id1,1],skel[id2,1]]
            zs=[skel[id1,2],skel[id2,2]]
    #         print xs,ys,zs
    #         print zs
            ax.plot(xs,ys,zs,color=plt.cm.gray(0) )
        ax.scatter(skel[:,0],skel[:,1],skel[:,2],c='green',s=100)

#         ax.set_xlim(-0.5, 0.5)
#         ax.set_ylim(-0.5, 0.5)
#         ax.set_zlim(-0.5, 0.5)
#         ax.set_xlabel('X axis')
#         ax.set_ylabel('Y axis')
#         ax.set_zlabel('Z axis')
plot_skel(new_skel_seq[1102])
```

```python
subject_id = 10
fall_ids = [0,1,2,3]
drawed = [False]*len(fall_ids)
        

for i in range(len(gt_seq)):
    lbl,subject = gt_seq[i]
    if subject == subject_id and lbl in fall_ids:
        if not drawed[fall_ids.index(lbl)] :
            fall_ids[fall_ids.index(lbl)] = True
            plot_skel(new_skel_seq[i],10)
        
```
